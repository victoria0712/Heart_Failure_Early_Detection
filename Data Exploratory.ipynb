```python
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn import utils
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import metrics
from sklearn.metrics import plot_precision_recall_curve, plot_roc_curve,f1_score,recall_score,precision_score
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv("heart_failure_clinical_records_dataset.csv")
```

```python
import plotly.figure_factory as ff
hist_data =[df["age"].values]
group_labels = ['age'] 

fig = ff.create_distplot(hist_data, group_labels)
fig.update_layout(title_text='Age distribution plot')

fig.show()
```

```python
import plotly.express as px
fig = px.box(df, x='sex', y='age', points="all")
fig.update_layout(
    title_text="Gender wise age spread - Male = 1 Female =0")
fig.show()
```

```python
import plotly.graph_objs as go
male = df[df["sex"]==1]
female = df[df["sex"]==0]

male_fail = male[df["Heart_failure"]==0]
male_not = male[df["Heart_failure"]==1]
female_fail = female[df["Heart_failure"]==0]
female_not = female[df["Heart_failure"]==1]

labels = ['Male - No heart failure','Male - Heart failure', "Female -  No heart failure", "Female - Heart failure"]
values = [len(male[df["Heart_failure"]==0]),len(male[df["Heart_failure"]==1]),
         len(female[df["Heart_failure"]==0]),len(female[df["Heart_failure"]==1])]
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.4)])
fig.update_layout(
    title_text="Heart Failure - Gender")
fig.show()
```

```python
fail = df[df["Heart_failure"]==0]["age"]
not_fail = df[df["Heart_failure"]==1]["age"]
hist_data = [fail,not_fail]
group_labels = ['Heart Failure', 'Not heart failure']
fig = ff.create_distplot(hist_data, group_labels, bin_size=0.5)
fig.update_layout(
    title_text="Analysis in age on different status")
fig.show()
```

```python
#analyse how smoking is related to heart disease
fig = px.violin(df, y="age", x="smoking", color="Heart_failure", box=True, points="all", hover_data=df.columns)
fig.update_layout(title_text="Analysis in age and smoking on heart failure")
fig.show()
```

```python
#The Heart Failure occur rate is high for not smoking person 55 to 65, while for smoking person it is between 50 to 60
#Death event for smoking person is high than not smoking person
```

```python
#explore the effect of diabetes
fig = px.violin(df, y="age", x="diabetes", color="Heart_failure", box=True, points="all", hover_data=df.columns)
fig.update_layout(title_text="Analysis in age and diabetes on heart failure")
fig.show()
```
